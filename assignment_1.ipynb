# Problem 1: Plotting the Club Members

For this problem, you need to write the function **plotData1** to plot the data scientists in the data-science club so that we can conveniently visualize their tenured years, salaries, and paid accounts or not.  Your plot should be similar to the figure that can be displayed below by running the code cell.


```python
x1 = [d[0] for d in data if d[2] == 0]
y1 = [d[1] for d in data if d[2] == 0]
x2 = [d[0] for d in data if d[2] == 1]
y2 = [d[1] for d in data if d[2] == 1]

def plotData1(data):
    plt.figure()
    plt.scatter(x1, y1, color='blue', marker='+', s=25, label='Unpaid')
    plt.scatter(x2, y2, color='green', marker='o', s=25, label='Paid')
    plt.title('Data-Science Club Members')
    plt.xlabel('Years as Data Scientist')
    plt.ylabel('Salary')
    plt.xlim(0, 12)
    plt.xticks(np.arange(0, 12, 1))
    plt.yticks(np.arange(20000, 120000, 20000))
    plt.legend(loc='upper left')
    plt.grid(True)
    plt.savefig('plot1', dpi=300)
    plt.show()
    pass

plotData1(data)
```

# Problem 2

For this problem, you need to write a function **plotData2** to create a histogram plot. The histogram plot allows us discover the underlying frequency distribution of tenure data associated with the data scientists including paid and unpaid members in the club.  Your plot should be similar to the figure that can be displayed below by running the code cell.

```python
data_paid = [d[0] for d in data if d[2] == 1]
data_unpaid = [d[0] for d in data if d[2] == 0]


def plotData2(data):
    plt.figure()
    plt.hist([data_unpaid, data_paid], bins=20, stacked=True, density=True, color=['blue', 'lightblue'], label=['Unpaid', 'Paid'])
    plt.title('Histogram of Tenure')
    plt.xlabel('Tenure Data')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.savefig('plot2', dpi=300)
    plt.show()
    pass

plotData2(data)
```

# Problem 3

For this problem, you need to write a function **plotData3** to make a figure with two subplots.  One is a bar plot, and the other is a histogram plot. The bar plot can conveniently support users to visualize how many accounts are paid and how many are unpaid. The specific numbers of accounts for each type should be presented. The histogram plot allows us discover the underlying frequency distribution of salary data associated with the two types of data scientists (paid and unpaid). # Problem 4 Community Data

For this problem, you need to write a function **barCommunityData** make a bar plot that shows numbers of counties in states.  Your plot should be similar to the figure that can be displayed below by running the code cell that has the Image object.  To approach this problem, you need to write the function definition that includes statements to parse the data from the provided file named "county.csv". For this assignment, you are {\bf not} allowed to use pandas library to process the data file.   You are required to read the file row by row and extract the data to address the question.  In your plot, the specific numbers of counties in the states should be presented.

For the reference result plot, it shows the numbers of counties in Arizona, California, and Texas. However, your **barCommunityData** applications should allow any other states on the second argument.  

```python
def barCommunityData(states):

    county_counts = {state: 0 for state in states}

    with open('county.csv', 'r') as file:
      next(file)
      for line in file:
            parts = line.strip().split(',')
            if parts[2] == '':
                continue
            state = parts[1]
            if state in county_counts:
                county_counts[state] += 1
            else:
                county_counts[state] = 1

      states_to_plot = [state for state in states if state in county_counts]
      counts = [county_counts[state] for state in states_to_plot]

      colors = ['blue', 'yellow', 'green', 'red', 'pink']

      plt.figure()
      bars = plt.bar(states_to_plot, counts, color=colors)
      plt.ylabel('Number of Counties')
      plt.title('Number of Counties per State')
      plt.grid(True)
      legend_labels = [f'{state}: {count} counties' for state, count in zip(states_to_plot, counts)]
      plt.legend(bars, legend_labels ,loc='upper left')
      plt.show()
    pass

barCommunityData(['Arizona', 'California', 'Texas'])
```
